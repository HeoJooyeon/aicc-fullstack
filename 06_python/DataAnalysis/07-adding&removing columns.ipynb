```python
# 평담 단가가 가장 높은 순서대로 다섯개의 정보를 알려주세요.
houses["price_sqft"] = houses["price"] / houses["sqft_living"] # (a-b).abs()
houses.sort_values("price_sqft", ascending = False).head()
```

```python
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm

plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['axes.unicode_minus'] = False

labels = ["최고단가","최저단가"]
values = [houses["price_sqft"].max(), houses["price_sqft"].min()]

# 최고가와 최저가를 볼 수 있도록 시각화 해주세요.
plt.bar(labels, values, color=["red","blue"])
plt.title("평당 단가 최고 vs 최저")
plt.ylabel("단가 ($ per sqft)")
plt.show()
```
