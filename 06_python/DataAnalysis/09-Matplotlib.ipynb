```python
import pandas as pd
titanic = pd.read_csv("data/titanic.csv")
plt.hist(titanic.age)
```

```python
titanic["age"] = pd.to_numeric(titanic["age"], errors="coerce")
plt.hist(titanic["age"],bins=20)
```

```python
plt.figure(figsize=(5,5))

plt.hist(titanic[titanic["pclass"] == 1]["age"], bins=20, color="yellow", alpha=0.9, label="1st Class")
plt.hist(titanic[titanic["pclass"] == 2]["age"], bins=20, color="#3498db", alpha=0.7, label="2nd Class")
plt.hist(titanic[titanic["pclass"] == 3]["age"], bins=20, color="#e74c3c", alpha=0.5, label="3rd Class")

plt.xlabel("Age")
plt.legend()
plt.title("Titanic Passenger Ages")
plt.show()
```

```python
# subplots
nums = np.arange(5) # [0, 1, 2, 3, 4]
plt.figure(figsize=(20,12))
plt.suptitle("Our First Subplot", fontsize=30)

plt.subplot(2,3,1)
plt.title("X")
plt.plot(nums, nums)

plt.subplot(2,3,2)
plt.plot(nums, nums ** 2, color="red")
plt.title("X Squared")

plt.subplot(2,3,3)
plt.plot(nums, nums ** 3, color="yellow")
plt.title("X Cubed")

plt.subplot(2,3,4)
plt.hist(titanic[titanic["pclass"] == 1]["age"], color="yellow", bins=20)
plt.title("1st Class")

plt.subplot(2,3,5)
plt.hist(titanic[titanic["pclass"] == 2]["age"], color="#3498db", bins=20)
plt.title("2nd Class")

plt.subplot(2,3,6)
plt.hist(titanic[titanic["pclass"] == 3]["age"], color="#e74c3c", bins=20)
plt.title("3rd Class")


plt.tight_layout()
plt.show()
```
