```python
# 여성 탐승자들의 생존여부를 그래프로 그려주세요
women = titanic["sex"] == "female"
titanic[women].survived.value_counts().plot(kind="pie")
```

```python
# 남성 탑승자들의 생존여여부를 그래프로 그려주세요
titanic[~women].survived.value_counts().plot(kind="pie")
```
