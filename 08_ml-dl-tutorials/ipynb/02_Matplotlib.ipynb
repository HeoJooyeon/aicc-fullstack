```python
t = np.random.randn(10000)
t
```

```python
plt.hist(t, bins=50, density=True)  # bins 막대 개수
plt.title('Histogram of Random Normal Data')
plt.xlabel('Value')  # x축 라벨
plt.ylabel('Frequency')  # y축 라벨
plt.grid(True)  # 격자 표시
plt.show()

# density=True 설명
# 기본값(density=False)은 각 구간에 속하는 데이터 개수(빈도)를 막대 높이로 표시합니다.
# density=True로 설정하면 전체 막대 아래 면적의 합이 1이 되도록 높이를 조정합니다. 즉, 확률 분포를 나타내는 밀도 함수처럼 표현됩니다.
```

```python
X = np.random.randn(100,2)
X
```

```python
# x축 좌표 (첫 번째 열),	y축 좌표 (두 번째 열)
plt.scatter(X[:,0], X[:,1])

plt.show()
```

```python

X1 = X[:50]  # 첫 50개
X2 = X[50:]  # 나머지 50개

plt.scatter(X1[:, 0], X1[:, 1], color='yellow')
plt.scatter(X2[:, 0], X2[:, 1], color='purple')


plt.show()
```

```python

X = np.random.randn(100,2)
colors = []
for x, y in X:
    if x < 0 and y < 0:
        colors.append('purple')
    else:
        colors.append('yellow')

plt.scatter(X[:, 0], X[:, 1], color=colors)
plt()
```

```python
X = np.random.randn(200, 2)      # 1) 평균이 0이고 표준편차가 1인 난수 200개 생성 (2차원)
X[:50] += 3                      # 2) 앞의 50개 데이터에 대해 (3,3)만큼 평행이동
Y = np.zeros(200)               # 3) 라벨 벡터 Y 생성 (200개 모두 0으로 초기화)
Y[:50] = 1                      # 4) 앞의 50개는 라벨(그룹을 나눔)을 1로 변경 → 클래스 1
plt.scatter(X[:, 0], X[:, 1], c=Y)  # 5) X의 점들을 Y값에 따라 색을 다르게
```
